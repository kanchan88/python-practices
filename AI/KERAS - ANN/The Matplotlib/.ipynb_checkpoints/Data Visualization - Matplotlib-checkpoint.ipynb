### Data Visualization using Matplotlib

1. plotting library for python
2. similar feel like matlab

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,10,11)
y = x*x
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT OREINTED
fig = plt.figure()
#add axes(left bata %, buttom bata %, height, width)
axes = fig.add_axes([0.1,0.1,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('FIRST')
axes[1].plot(y,x)
```

```python
fig
```

```python
fig.savefig('my_data.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Sqaured")
ax.plot(x,x**3, label="X Cubed")

ax.legend()
```

## PLOT APPREANCE

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="#101010", linewidth=2.5, linestyle='steps', marker='o') #RGB Hex Code 
# color="green" also possible
#alpha is for transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, markersize=10, marker='o',
       markerfacecolor='green')
```
