```python
engine = create_engine('postgresql://postgres:admin@localhost:5432/products')
data.to_sql('details', engine, if_exists='replace')
```

```python
#reading saved data
fromdb= pd.read_sql("select categories, count(categories) from details group by categories order by count(categories) desc limit 10", engine.connect());
fromdb.sort_values(by=['count'])
```
